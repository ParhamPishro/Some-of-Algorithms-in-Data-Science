```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.datasets import load_wine
```

**Sample**

```python
wine = load_wine()
```

```python
wine_df = pd.DataFrame(wine.data , columns=wine.feature_names)
wine_target=pd.DataFrame(wine.target , columns=['target'])
wine_df.head()
```

```python
tsne = TSNE()
X_embedded = tsne.fit_transform(wine_df)
```

```python
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], hue=wine_target['target'], legend='full')
```
