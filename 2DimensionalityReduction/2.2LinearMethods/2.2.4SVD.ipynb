```python
#importe libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.linalg import svd
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
```

```python
#read data

from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_excel('/content/gdrive/My Drive/DataSet/Escaperoom.xlsx')
```

```python
features = df.values[:,:9]
target = df.values[:,10:]
```

```python
# define a matrix
A = features
print("A=", A)

# SVD
U, S, V_T = svd(A)

# left singular vectors
print("U=", U)

# singular values
print("S=",S)

#right singular vectors
print("V_T", V_T)
```

```python
print(A.shape)
print(U.shape)
print(S.shape)
print(V_T.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(range(9), S, alpha=0.5, align='center', label='S')
plt.legend()
plt.ylabel('S')
plt.xlabel('SVD')
plt.show()
```

1 Singular Value

```python
#Fitting the SVD class

A1 = csr_matrix(features)
svd1 = TruncatedSVD(n_components=1, n_iter=7, random_state=42)
svd1.fit(A1)
TruncatedSVD(n_components=1, n_iter=7, random_state=42)
```

```python
svd1.singular_values_
```

```python
svd1.explained_variance_ratio_
```

2 Singular Value

```python
#Fitting the SVD class

A2 = csr_matrix(features)
svd2 = TruncatedSVD(n_components=2, n_iter=7, random_state=42)
svd2.fit(A2)
TruncatedSVD(n_components=2, n_iter=7, random_state=42)
```

```python
svd2.singular_values_
```

```python
svd2.explained_variance_ratio_
```

```python
svd2.explained_variance_ratio_.sum()
```
